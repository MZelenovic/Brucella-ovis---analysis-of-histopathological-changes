# Statistics

First, we will check whether the histopathological parameters are statistically significantly different in PCR positive and PCR negative organs

Based on these results and the threshold value alpha = 0.05, we can conclude that there is no statistically significant difference between PCR positive and PCR negative organs for any of the histopathological parameters.

```python
from scipy.stats import mannwhitneyu

'''Null hypothesis (H0): Data from PCR positive organs do not differ from data from PCR negative organs.
Alternative hypothesis (H1): There is a significant difference between the data of PCR positive organs and PCR negative organs.'''

'''In our case, we have numerical data ranging from 0 to 4 and two groups (PCR positive and PCR negative).
In order to test the hypothesis of differences between these groups, we can use a non-parametric test for independence
samples, such as the Mann-Whitney U test. This test is used to compare the distribution of numerical data between
two independent groups.'''

statistic, p_value = mannwhitneyu(pcr_poz, pcr_neg)

# printing results
print(f"Statistics Mann-Whitney U: {statistic}")
print(f"P-value: {p_value}")

# Interpretation of results
alpha = 0.05  # Level of significance
```

```python
# data visualization

for feature in pcr_poz.columns:
    plt.figure(figsize=(8, 6))
    plt.hist(pcr_poz[feature], alpha=0.5, label='PCR positives')
    plt.hist(pcr_neg[feature], alpha=0.5, label='PCR negatives')
    plt.title(f'Histogram for {feature}')
    plt.legend()
    plt.show()
```

```python
from scipy.stats import ttest_ind

alpha = 0.05  # # Level of significance

for feature in pcr_poz.columns:
    t_statistic, p_value = ttest_ind(pcr_poz[feature], pcr_neg[feature])

    # printing results
    print(f"Feature: {feature}")
    print(f"Statistics od t-test: {t_statistic}")
    print(f"P-value: {p_value}")

    # Interpretation of results
    if p_value < alpha:
        print("There is a statistically significant difference between the groups.")
    else:
        print("There is no statistically significant difference between the groups.")

    print("\n")
```

### Comparative analysis of the intensity of changes in paired organs

With epididymitis in rams, it is a common phenomenon that the changes in one testicle and epididymis are more pronounced. We will check if this is true in our case as well.

```python
# Adding a column of summarized parameter values for each organ
hp['Sum_scor'] = hp.iloc[:, 3:16].sum(axis=1)
hp.head()
```

```python
# Division of the set into one in which the changes on the organs are less pronounced and the other with more intense changes
hp_small = hp[hp.groupby('Ram')['Sum_scor'].transform('min') == hp['Sum_scor']]
hp_large = hp[hp.groupby('Ram')['Sum_scor'].transform('max') == hp['Sum_scor']]
hp_large = hp_large.drop_duplicates(subset=['Ram', 'Sum_scor'])
hp_small = hp_small.drop_duplicates(subset=['Ram', 'Sum_scor'])
hp_large.head()
```

```python
# Checking the normality of differences between paired samples

# Calculation of differences between samples
diff = hp_large['Sum_scor'].to_numpy() - hp_small['Sum_scor'].to_numpy()
# Setting the graph size and style
plt.figure(figsize=(6, 5))
sns.set(style="whitegrid")

# Drawing a histogram
sns.histplot(diff, kde=True, color='skyblue')
plt.title('')
plt.xlabel('The difference in the intensity of changes')
plt.ylabel('Number of samples ')
```

```python
# Statistical test (Shapiro-Wilk Test)
from scipy.stats import shapiro

# Shapiro-Wilk test
stat, p_value = shapiro(diff)
# Ispis rezultata
print(f'Statistics: {stat}\nP-value: {p_value}')

# Normality check
alpha = 0.05
if p_value < alpha:
    print("The differences are not normally distributed.")
else:
    print("There is insufficient evidence that the differences are not normally distributed.")
```

Since we checked that the condition of normality of the distribution of differences between paired samples is met, we can use the T test.

```python
from scipy.stats import ttest_ind, ttest_rel

'''Null hypothesis (H0): There is no statistically significant difference between the mean values of the groups.
Alternative hypothesis (H1): There are statistically significant differences between the mean values of the groups.'''

# Paired t-test
stat, p_value = ttest_rel(hp_small['Sum_scor'], hp_large['Sum_scor'])

# Print the results
print(f'T-statistics: {stat}\nP-value: {p_value}')

# Testing for statistical significance
alpha = 0.05
if p_value < alpha:
    print("There are statistically significant differences between the groups.")
else:
    print("There is insufficient evidence that there are statistically significant differences.")
    
# Independent t-test
stat, p_value = ttest_ind(hp_small['Sum_scor'], hp_large['Sum_scor'])

# Print the results
print(f'T-statistics: {stat}\nP-value: {p_value}')

# Testing for statistical significance
alpha = 0.05
if p_value < alpha:
    print("There are statistically significant differences between the groups.")
else:
    print("There is insufficient evidence that there are statistically significant differences.")
```

In the case of paired data, where we have measurements on the same animals (in this case, testes and epididymis), it is recommended to use the t-test for paired samples. In a paired-samples t-test, each data in one sample is directly related to the corresponding data in the other sample. For the paired-samples t-test, it is assumed that the differences between the paired samples are normally distributed. Since this assumption is satisfied and the p-value is low (0.0000095), we can conclude that there are statistically significant differences between the groups.

### ELISA


We will examine whether there is a statistically significant difference between the number of affected males and females (although it is immediately obvious that the difference is huge).

```python
from scipy.stats import fisher_exact

# Creating contingency table
# We have 23 positive and 10 negative males and 2 positive and 59 negative females
table = [[23, 10], [2, 59]]

# Fischer's exact test
odds_ratio, p_value = fisher_exact(table)

# Print the results
print(f'Odds Ratio: {odds_ratio}\nP-value: {p_value}')
```

The obtained results of Fischer's exact test indicate that there is a statistically significant difference between males and females in terms of positive results on the serological test. The interpretation is as follows:

Odds Ratio: Odds ratio of 67.85 indicates that the chances of a positive result in males are almost 68 times higher than in females.

P-value: The P-value of 4.19e-12 (0.00000000000419) is extremely low. This low p-value suggests that the difference in positive results between males and females is very likely real and cannot be attributed to chance.

Considering these results, we can conclude that there is a statistically significant difference in the response to the serological test between males and females. These results support the hypothesis that there is a difference between the two sexes in the context of the serological test we conducted.

### PCR sensitivity and specificity

We will calculate the sensitivity and specificity of the PCR test for the samples of testicles and epididymis of males, because only there we had positive results. We had two extraction kits, Qiagen with 6 positive and Thermo Scientific with 4 positive samples. From the data we have, we should estimate sensitivity (the ability of the test to identify true positive cases) and specificity (the ability of the test to identify true negative cases). We will take into account that we cannot accurately calculate the number of really positive and negative cases due to the lack of complete data. We will assume that ELISA is our "reference" test. According to the manufacturer, the Ingezim Brucella ovis kit has a sensitivity of 98.5% compared to isolation and 99% compared to RVK. Based on these assumptions, the sensitivity of the PCR test can be calculated using the following formula: Sensitivity (True Positive Rate) = Total number of positive samples according to the PCR test / Number of true positive ELISA results

The specificity of the PCR test can be calculated using the formula: Specificity (True Negative Rate) = Total number of negative samples according to the PCR test / Number of truly negative ELISA results. Considering that we have no negative ELISA results, we are not able to calculate the specificity (dividing by 0), we will only calculate the total number of negative results

```python
number_of_true_positive_PCR_qiagen = 6
number_of_true_positive_PCR_TS = 4
total_number_of_positive_ELISA = 38

# Calculation of sensitivity and specificity for males only
Sensitivity_qiagen = number_of_true_positive_PCR_qiagen / total_number_of_positive_ELISA
Sensitivity_TS = number_of_true_positive_PCR_TS / total_number_of_positive_ELISA
# Calculation of false negatives
False_negatives_qiagen = total_number_of_positive_ELISA - number_of_true_positive_PCR_qiagen
False_negatives_TS = total_number_of_positive_ELISA - number_of_true_positive_PCR_TS

# Print the results
print(f'Sensitivity of the PCR assay for the Qiagen kit: {Sensitivity_qiagen:.2%}')
print(f'False negative PCR results for the Qiagen kit: {False_negatives_qiagen}')
print(f'\nSensitivity of the PCR assay for the Thermo Scientific kit: {Sensitivity_TS:.2%}')
print(f'False negative PCR results for the Thermo Scientific kit: {False_negatives_TS}')
```

### Now we will examine whether the histopathological changes are more pronounced in the testicles or epididymis. Considering that the disease is called eipdidymitis of rams, we expect that the changes are more intense in the flocks.

```python
'''Null hypothesis (H0): There is no statistically significant difference between the histopathological parameters of
the testis and epididymis.
Alternative hypothesis (H1): There are statistically significant differences between the histopathological parameters of
the testisand epididymis.'''

# data distribution check

# Visualization of data distribution - Histogram
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.hist(hp['Fibrosis testis'], bins=10, color='blue')
plt.title('Fibrosis testis')

plt.subplot(1, 2, 2)
plt.hist(hp['Fibrosis epididimis'], bins=10, color='green')
plt.title('Fibrosis epididimis')

plt.tight_layout()
plt.show()
```


Since the distribution is not normal, we will use the Wilcoxon paired test, which is not overly sensitive to assumptions about the data distribution.

```python
from scipy.stats import wilcoxon

stat, p_value = wilcoxon(hp['Fibrosis testis'], hp['Fibrosis epididimis'])

print(f"Statistcs: {stat:.2f}")
print(f"P-value: {p_value:.6f}")
```


The Wilcoxon test warns us that the calculation may not be good because we have zero values in our data. We will also try the Mann-Whitney U test, which is also not sensitive to assumptions about the data distribution and may be more resistant to the presence of zeros.

```python
stat, p_value = mannwhitneyu(hp['Fibrosis testis'],hp['Fibrosis epididimis'])

print(f"Statistics: {stat:.2f}")
print(f"P-value: {p_value:.6f}")
```

The p-values of 0.03 (Wilcoxon) and 0.043 (Mann-Whitney U) are less than the usual significance level of 0.05, indicating that there is a statistically significant difference between these two data samples. This suggests that the statistical distribution of values in "Fibrosis Testis" and "Fibrosis Epididymis" differs, although this difference may be small, it is considered significant at the 0.05 level.

Additionally, the statistic value of 77.00 (Wilcoxon) and 543.00 (Mann-Whitney U) are the test statistic values used to evaluate these differences.

This indicates that there is a basis for rejecting the null hypothesis (which assumes no difference between the groups) in favor of the alternative hypothesis, which means that it can be assumed that there is a real difference between the intensity of fibrotic changes in seminiferous tubules and seminiferous tubules.
