```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('epididymis cluster analysis.xlsx')
df.drop(columns=['Organ-epididimis','sum score', 'hijer. Div', 'hijer. Aggl',
       'k-means'], inplace=True)
df.columns
```

```python
df.head()
```

```python
# Grouping of data by clusters
cluster_groups = df.groupby('klaster')

#Calculation of average values for each variable by clusters
average_values = cluster_groups.mean()
average_values
```

```python
# Ploting a Boxplpt

# data transformation
df_melted = df.melt(id_vars='klaster', var_name='Changeable', value_name='Value')

# Podešavanje veličine grafičkog prikaza
plt.figure(figsize=(16, 8))

# Prikazivanje boxplotova za sve promenljive sa oznakama klastera
sns.boxplot(data=df_melted, x='Value', y='Changeable', hue='klaster', palette='Set2')
plt.xlabel('Value',size=16)
plt.ylabel('Changeable',size=16)
#plt.title('Расподјела промјенљивих по Кластерима', size=20)
plt.legend(title='Klaster', fontsize='large', title_fontsize='large')
plt.show()
```

```python
# Performin ANOVA
from scipy.stats import f_oneway

# extracting data for each cluster
klaster_1 = df[df['klaster'] == 1]
klaster_2 = df[df['klaster'] == 2]

# Application of ANOVA test for each variable
for column in df.columns[:-1]:
    stat, p_value = f_oneway(klaster_1[column], klaster_2[column])
    print(f"ANOVA results for the variable '{column}':")
    print(f"F statistics: {stat}")
    print(f"P-value: {p_value}")
    print()
```

There are statistically significant differences for all changes between different clusters except for the age of rams

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Defining data for analysis
data = df[['Age', 'Infiltr. Ly, Pc, Mf', 'Degeneration/vakuolisation epitel',
       'Inflammatory cells (epithelium)',
       'Dilatation of the epididymis canaliculi ', 'Hyperemia',
       'Interstitial edema', 'Fibrosis', 'Sperm', 'klaster']]

# Filtriranje podataka za grupe 1 i 2
group1 = data[data['klaster'] == 1]
group2 = data[data['klaster'] == 2]

# Display of Tukey's test results for each variable
for column in data.columns[:-1]:  # Ignore the 'cluster' column
    result = pairwise_tukeyhsd(data[column], data['klaster'])
    print("Tukey's test for a variable:", column)
    print(result)
    print()
```

Tukey's post-hoc test was used after ANOVA analysis to identify specific differences between groups when ANOVA analysis indicated that there was a statistically significant difference between groups.

This test allows us to determine exactly which groups have significant differences in means. In the test results, we get information about the difference in means between each possible combination of groups, along with the lower and upper bounds of the confidence interval for the difference, and whether the differences are statistically significant.

Therefore, we can conclude that there is a statistically significant difference between clusters 1 and 2 for all the variables we analyzed except for age. This further confirms that there is heterogeneity in pathological processes between these two groups.

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score
from scipy.spatial.distance import pdist, squareform

# Data extraction for analysis
X = df.iloc[:, :-1]
labels = df['klaster']

# 1. Silhouette indekx
silhouette_avg = silhouette_score(X, labels)

# 2. Hartigan indekx (approximation)
intra_cluster_distances = np.mean([np.mean(pdist(X[labels == label], metric='euclidean')) for label in set(labels)])
centroids = X.groupby(labels).mean()
inter_cluster_distances = pdist(centroids, metric='euclidean')
hartigan_index = np.mean(intra_cluster_distances) / np.mean(inter_cluster_distances)

# 3. Calinski & Harabasz index
ch_score = calinski_harabasz_score(X, labels)

# Printing results
print(silhouette_avg)
print(hartigan_index)
print(ch_score)
```

### Silhouette index (0.2719):

The value of the Silhouette index varies from -1 to 1. Values ​​closer to 1 indicate that the clusters are well separated, while values ​​closer to 0 indicate overlapping clusters. Given our value of 0.2719, we can conclude that there is some separation between clusters, but there may be room for improvement in terms of better separation.

### Hartigan index (0.9900):

The Hartigan index measures cluster density. Values ​​closer to zero indicate that the clusters are well defined and dense. Our value of 0.9900 suggests that the clusters are relatively dense, which is a good sign.

### Calinski & Harabasz index (17.7665):

The Calinski & Harabasz index measures cluster compactness and separation between clusters. Higher values ​​of this index indicate more compact clustering. Your value of 17.7665 suggests that there is relatively good compact clustering, but there could be room for improvement, especially regarding the separation between clusters. In essence, these results indicate that there is some separation between clusters, but compactness and separation may need to be improved to achieve better clustering.
